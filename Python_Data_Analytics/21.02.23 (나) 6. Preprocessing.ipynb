# 3. Filtering(필터링)
> ## 1) 실습용 'titanic' 데이터셋

```
import seaborn as sns
titanic = sns.load_dataset('titanic')

titanic.head(3)
```

> ## 2) 'age'가 10살 이상이면서 20살 미만

```
filter_1 = (titanic.age >= 10) & (titanic.age < 20)
titanic.loc[filter_1, :].head()
# filter_1 조건에 맞는 행과 모든 열을 출력
```

> ## 3) 'age'가 10살 미만이면서 'sex'이 여자

```
filter_2 = (titanic.age < 10) & (titanic.sex == 'female')
titanic.loc[filter_2, :].head()
```

> ## 4) 'age'가 10살 미만 또는 20살 이상

- 'age', 'sex', 'alone' 열만 출력

```
filter_3 = (titanic.age < 10) | (titanic.age >=20)
titanic.loc[filter_3, ['age','sex','alone']].head()
```

> ## 5) isin( ) : 열이 list의 값들을 포함하고 있는 모든 행들을 뽑아 냄

- 'sibsp'에 3 또는 4 또는 5를 포함
   - 'sipsp' : 함께 탑승한 형제 또는 배우자의 수
   - 'sibsp ==3 | sibsp == 4 | sibsp == 5'

```
filter_isin = titanic['sibsp'].isin([3, 4, 5])
titanic[filter_isin].head(6)
# sibsp 값이 3,4,5인 모든 행을 가져오되 6행만 보여줌
```

```
titanic['sibsp']
```

# 4. 데이터프레임 합치기
> ## 1) 데이터프레임 df1, df2 생성

```
import pandas as pd
df1 = pd.DataFrame({'HP': ['hp0', 'hp1', 'hp2', 'hp3'],
                    'IBM' : ['ibm0', 'ibm2', 'ibm2', 'ibm3'],
                    'DELL' : ['dell0', 'dell1', 'dell2', 'dell3']},
                   index = [0, 1, 2, 3])
df1
```

```
df2 = pd.DataFrame({'HP': ['hp2', 'hp3', 'hp4', 'hp5'],
                    'IBM' : ['ibm2', 'ibm3', 'ibm4', 'ibm5'],
                    'DELL' : ['dell2', 'dell3', 'dell4', 'dell5'],
                    'ASUS' : ['asus2', 'asus3', 'asus4', 'asus5']},
                   index = [2, 3, 4, 5])
df2
```

> ## 2) concat( ) : 연결

- 행기준 : axis = 0 
   - 열 개수는 그대로, 행 늘어남 (상 + 하)

```
pd.concat([df1,df2], axis = 0)
```

- ignore_index = True

```
pd.concat([df1, df2], axis = 0, ignore_index = True)
# 기존 index 무시하고 자동으로 부여
```

- 열기준 : axis = 1, join = 'inner'
   - 'inner' : Intersection
      - 공통 행만 추출, 열은 좌 + 우 연결

```
pd.concat([df1, df2], axis = 1, join = 'inner')
```

- 열기준 : axis = 1, join = 'outer'
   - 'inner' : Union
      - 모든 데이터 연결, 상 + 하, 좌 + 우

```
pd.concat([df1, df2], axis = 1, join = 'outer')
```

> ## 3) merge( ) : 덮음

- how = 'inner'
   - 공통 key를 가진 행만 출력하되, 오른쪽에서 왼쪽으로 DF 덮은 열 출력

```
pd.merge(df1, df2)
```

-  how = 'outer'
   - 모든 데이터 연결, 오른쪽에서 왼쪽으로 DF 덮음

```
pd.merge(df1, df2, how = 'outer')
```

- how = 'left'
   - 왼쪽 데이터프레임의 키(Key)만 사용 'left-outer'
   - 오른쪽 열로 덮기 때문에 왼쪽에는 있지만 오른쪽에는 없는 데이터는 결측값으로 표시

```
pd.merge(df1, df2, how = 'left')
```

- how = 'right'
   - 오른쪽 데이터프레임의 키(Key)만 사용 'right-outer'
   

```
pd.merge(df1, df2, how = 'right')
```

> ## 3) agg( )

- Aggregation : 여러 개의 함수를 groupby 객체에 적용
   - 그룹별로 연산 결과를 집계하여 반환

```
grouped.agg({'fare' : ['min', 'max'], 'age' : ['mean', 'std']})
# class별로 fare(표값)는 최소, 최대값을, age(나이)는 평균, 표준편차 출력
```

> ## 4) transform( )

- 그룹별로 함수를 적용하여 각 원소의 행과 열을 기준으로 연산 결과를 반환
   - 데이터프레임에 'z_score'열(Column)을 추가

- 원본 DF

```
df.head(3)
```

- z_score( ) 표준화 함수 정의

```
def z_score(x) :
  return (x - x.mean()) / x.std()
```

- DF에 transform( )함수를 적용하여 'z_score' 열 추가
   - 3개 그룹별 'age'변수에 대한 표준화 변수('z_score') 계산

```
df['z_score'] = grouped.age.transform(z_score)
# 데이터프레임df에 age변수에 z_score함수를 적용한 'z_score'열 새로 만듬
```

- 추가된 'z_score'열 확인

```
df.head(3)
```

# 7. pivot_table( )

> ## 1) 실습용 'titanic' 데이터셋

```
import seaborn as sns
titanic = sns.load_dataset('titanic')

df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
df.head(3)
```

> ## 2) pivot_table( ) 구성요소 : 이원분할표

- index : 행 인덱스
- column : 열 인덱스
- values : 데이터
- aggfunc : 적용 함수

```
# 이원분할표
df_1 = pd.pivot_table(df,
                      index = 'class',
                      columns = 'sex',
                      values = 'age',
                      aggfunc = 'mean')
df_1
# 행 : class / 열 : sex / 데이터값 : 평균 나이
```

> ## 3) 두 개의 적용 함수

```
df_2 = pd.pivot_table(df,
                      index = 'class',
                      columns = 'sex',
                      values = 'survived',
                      aggfunc = ['mean', 'sum'])
df_2
# 행 : class / 열 : sex / 데이터값 : 평균 생존율과 생존자 수
```

> ## 4) 다중 인덱스, 다중 데이터, 다중 함수 : 삼원 분할표

```
df_3 = pd.pivot_table(df,
                      index = ['class','sex'],
                      columns = 'survived',
                      values = ['age','fare'],
                      aggfunc = ['mean', 'max'])
df_3
# 행 : class, sex / 열 : 생존여부(0,1) / 데이터값 : 사망자, 생존자의 평균 나이,표값과 최대 나이, 최대 표값
# 해석 : 낮은 등급칸의 여자 생존자 나이가 높은 등급칸 나이보다 평균적으로 어리다.
```

```
df_3.xs('mean', axis = 1)
# 모든 칸, 성별의 평균 나이, 표값
```

- 행 멀티인덱스 : 객실등급이 일등실이면서 성별이 남자
