# Scaling
- 범위(Scale)가 다른 변수의 범위(Scale)를 비슷하게 맞추기 위한 목적
# 1. 실습 데이터
> ## 1) seaborn 'mpg' Date Set

```
import seaborn as sns
df = sns.load_dataset('mpg')
```

- 자동차 연비(mpg) 데이터
   - mpg : miles per gallon
   - 1 mile : 1,6 Km
   - 1 gallon : 3.78 Liter

```
df.info()
# horsepower에 결측값이 있지만, weight, mpg만 변수로 사용할 것이기 때문에 상관 없다.
```

```
df.head()
```

```
X = df[['weight']]  # DF
y = df['mpg']       # Series
```

> ## 2) Without Scaling

- X값 확인

```
X[:5]
```

```
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X, y = y)
plt.xlabel('weight_Without_Scaling')
plt.show()
```

> ## 3) With Normalization

- sklearn Package

```
from sklearn.preprocessing import MinMaxScaler
scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

- 정규화된 X값 확인

```
X_Norm[:5]
```

- 정규화된 X값 모델생성

```
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Norm, y = y)
plt.xlabel('weight_With_Normalization')
plt.show()
```

> ## 4) With Standardization

- skleaarn Package

```
from sklearn.preprocessing import StandardScaler
scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

- 표준화된 X값 확인

```
X_Stan[:5]
```

-  표준화된 X값 모델 생성

```
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Stan, y = y)
plt.xlabel('weight_With_Standardization')
plt.show()
```
