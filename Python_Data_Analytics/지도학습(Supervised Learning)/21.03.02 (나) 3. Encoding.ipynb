# 1. 실습 데이터
> ## 1) seaborn 'mpg' Data Set

```
import seaborn as sns
df = sns.load_dataset('mpg')
```

```
df.info()
```

- 문자형 데이터 : 'origin'

```
df.head()
```

```
type(df.origin[0]) # 변수 타입을 알고 싶을 땐, [0]붙이고, DF인지 Series인지 알고 싶으면 안 붙여도 됨
```

- 명목형 : 이름 확인 및 빈도분석

```
df.origin.value_counts()
```

- 'origin'Data

```
X = df[['origin']]
```

```
X[111:115]
```

> ## 2) With LabelEncoder

- 정수(Integer) 인코딩

```
from sklearn.preprocessing import LabelEncoder
encoder1 = LabelEncoder()
LE = encoder1.fit_transform(X)
```

- 정수 인코딩 결과

```
LE[111:115] 
# 알파벳 순서대로, europe = 0, japan = 1, usa = 2
```

> ## 3) With OneHotEncoder : 가변수

```
from sklearn.preprocessing import OneHotEncoder
encoder2 = OneHotEncoder()
OHE = encoder2.fit_transform(X)
```

```
type(OHE)
```

- Array 변환 필요

```
print(OHE[111:115])
```

```
OHE.toarray()[111:115]
```
