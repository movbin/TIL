# Gradient Descent with Error

- Error = (y - y_hat)

# 1. 함수 정의

> ## 1) Import numpy

```
import numpy as np
```

> ## 2) sigmoid( ) 함수 정의

```
def sigmoid(x):
  y_hat = 1 / (1 + np.exp(-x))
  return y_hat
```

> ## 2) 'x', 'y', 'w', 'b' 설정

- 'input', 'output' 정의
- 'w', 'b' 초기화

```
x = 0
y = 1

w = 0.9
b = 0.2
```

```
for i in range(1501):
  y_hat = sigmoid(x * w + b)
  error = y - y_hat

  w = w + x * 0.1 * error
  b = b + 1 * 0.1 * error

  if i % 100 == 0:
    print(i, error, y_hat)
# 수치미분을 하는 경우보다 속도가 훨씬 빠르다
```
