# Ⅰ.MNIST Data_Set Load & Review


> ## 1) Load MNIST Data_Set

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

- Train_Data Information

```python
print(len(X_train))
print(X_train.shape) # 이미지 6만장이 28 x 28의 일정한 크기로 구성되어 있다

print(len(y_train))
print(y_train[0:5])
```

- Test_Data Information

```python
print(len(X_test))
print(X_test.shape) # 이미지 1만장이 28 x 28의 일정한 크기로 구성되어 있다


print(len(y_test))
print(y_test[0:5])
```

> ## 2) Visualization

```python
import matplotlib.pyplot as plt
digit = X_train[4]  # 9인지 확인
plt.imshow(digit, cmap = 'gray')
plt.show()
```

```python
import numpy as np
np.set_printoptions(linewidth = 150) 

print(X_train[4]) # array로 확인
```

# Ⅱ. Data Preprocessing



> ## 1) Reshape and Normalization

- reshape
   - (60000, 28, 28) to (60000, 28 * 28)
   - 한 차원 내린다

```python
X_train = X_train.reshape((60000, 28 * 28))
X_test = X_test.reshape((10000, 28 * 28))

X_train.shape, X_test.shape
```

- Normalization

```python
X_train = X_train.astype(float) / 255
X_test = X_test.astype(float) / 255
# 정규화식 : x - min / max - min인데, min = 0, max = 255이라서 255로 나눈다
```

```python
print(X_train[4])
```

> ## 2) One Hot Encoding

```python
from keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
print(y_train[:5])
# 5, 0, 4, 1, 9 표현

print(y_test[:5])
# 7, 2, 1, 0, 4 표현
```

# Ⅲ. MNIST Keras Modeling

> ## 1) Model Define

- 모델 신경망 구조 정의
   - 2개의 Hidden Layers & 768개의 nodes
   - 복잡한 model capacity로 인한 overfitting

```python
from keras import models
from keras import layers

mnist = models.Sequential()
mnist.add(layers.Dense(512, activation = 'relu', input_shape = (28 * 28,))) # 784 * 512 + 512
mnist.add(layers.Dropout(0.4))
mnist.add(layers.Dense(256, activation = 'relu')) # 512 * 256 + 256
mnist.add(layers.Dropout(0.2))
mnist.add(layers.Dense(10, activation = 'softmax')) # 256 * 10 + 10
```

- 모델 구조 확인

```python
mnist.summary()
# 파라미터가 53만개 그대로.
```

> ## 2) Model Compile

- 모델 학습방법 설정

```python
mnist.compile(loss = 'categorical_crossentropy',
              optimizer = 'rmsprop',
              metrics = ['accuracy'])
# 0 ~ 9를 분류해야 하니까 다중분류_CEE
```

> ## 3) Model Fit

- 약 3분

```python
%%time
hist_mnist = mnist.fit(X_train, y_train,
                       epochs = 100,
                       batch_size = 128,
                       validation_split = 0.2)
```

> ## 4) 학습 결과 시각화 - Overfitting

- Loss Visualization

```python
import matplotlib.pyplot as plt
epochs = range(1, len(hist_mnist.history['loss']) + 1)

plt.figure(figsize = (9, 6))
plt.plot(epochs, hist_mnist.history['loss'])
plt.plot(epochs, hist_mnist.history['val_loss'])
plt.title('Training & Validation Loss')
# plt.ylim(0, 0.2)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Training Loss', 'Validation Loss'])
plt.grid()
plt.show()
# overfitting이 기존모델보다 줄었음을 확인
```

> ## 5) Model Evaluate

- Loss & Accuracy

```python
loss, accuracy = mnist.evaluate(X_test, y_test)

print('Loss = {:.5f}'.format(loss))
print('Accuracy = {:.5f}'.format(accuracy))
# test data에서 accuracy = 0.984로 가장 높게 나옴
# Accuracy = Dropout모델 > 기본 모델 > 파라미터 줄인 모델 > 규제화 + 파라미터 줄인 모델 > 규제화
```

> ## 6) Model Predict

- Probability

```python
np.set_printoptions(suppress = True, precision = 9)
print(mnist.predict(X_test[:1,:]))
```

- Class

```python
print(mnist.predict_classes(X_test[:1,:]))
```
