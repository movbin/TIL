- DataFrame 형태도 진행된다.
   - Hanspell은 Series로 진행 됐음

```python
# Hanspell 진행 후에도 띄어쓰기가 안 되어 있는 문장만 뽑아 
# PyKoSpacing(띄어쓰기 패키지) 진행한다. 
%%time

from pykospacing import spacing

spacing_sentence7 = []

for s in sentence7['7']:
    if ' ' not in s:                    # 띄어쓰기가 없다면
        print(s)                        # 해당 문장을 뽑아서
        kospacing_sentence = spacing(s) # pykospacing으로 띄어쓰기 해라
        spacing_sentence7.append(kospacing_sentence)
    else:
        spacing_sentence7.append(s)      # 띄어쓰기가 되어 있다면, 그대로 spacing_sentence에 추가해라
```
