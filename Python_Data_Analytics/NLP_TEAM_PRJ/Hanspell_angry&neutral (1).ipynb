```python
sentence = pd.DataFrame() # 빈 데이터프레임 생성
```

```python
sentence['original'] = df['Sentence'] # 불러온 엑셀에서 문장만 sentence 데이터프레임의 original 열에 추가

sentence.head()
```

```python
# 문장의 앞뒤 공백을 제거한 new_sentence를 만들자
new_sentence = []

for s in sentence['original']:
    new_sentence.append(s.strip())
```

```python
new_sentence
```

```python
sentence['new'] = pd.DataFrame(new_sentence)  # 공백 제거한 new_sentence를 sentence df에 new가 열로 추가
```

```python
sentence.head()
```

```python
# 희진님이 정답 데이터 있는 데이터로 실험해주신 '가장 정확도가 높은 버전'으로 데이터를 변경
# 정규방정식을 통해 바꾼다.

# 정확도가 높은 버전 = ♡를 제외한 특수문자만 제거한, 원본 데이터를 최대한 살린 상태로 진행
withoutmark = []
withoutmark = sentence['new'].str.replace('[^가-힣A-Za-z0-9♡]', '') 

# 이렇게 써도 열이 추가 되는 것인지 써본다. ㅇㅇ된다. 
# sentence['hanspell_without_mark'] = sentence['new'].str.replace('[^가-힣 ]', '') # 한글 외 문자 및 자음, 모음 제거
```

```python
sentence['without_mark'] = pd.DataFrame(withoutmark)
```

```python
sentence.head()
```

```python
# without_mark열로 hanspell 패키지 적용
%time

from hanspell import spell_checker

hanspell_result = []

for s in sentence['without_mark']:  
    hanspell = spell_checker.check(s) # hanspell 활용하여 띄어쓰기 및 맞춤법 검사
    hanspell_result.append(hanspell.checked)
```

```python
# Hanspell 진행 후에도 띄어쓰기가 안 되어 있는 문장만 뽑아 
# PyKoSpacing(띄어쓰기 패키지) 진행한다. 
%%time

from pykospacing import spacing

spacing_sentence = []

for s in sentence['hanspell_result']:
    if ' ' not in s:                   # 띄어쓰기가 없다면
        print(s)
        kospacing_sentence = spacing(s) # pykospacing으로 띄어쓰기 해라
        spacing_sentence.append(kospacing_sentence)
    else:
        spacing_sentence.append(s)      # 띄어쓰기가 되어 있다면, 그대로 spacing_sentence에 추가해라
```

```python
sentence['spacing_result'] = pd.DataFrame(spacing_sentence)
```
