# 5. Casting : 바꾼다
> ## 1) Data Type

- int to float : 정수를 실수로

```
print(type(9))
print(type(float(9)))
```

- str to float : 문자를 실수로

```
print(type('9.4'))
print(type(float('9.4')))
```

- float to int : 실수를 정수로

```
print(type(9.0))
print(9.0)

print(type(int(9.0)))
print(int(9.0))
```

- str to int : 문자를 정수로

```
print(type('9'))
print(9)

print(type(int('9')))
print(float(9))
```

- float to int : 실수를 정수로
- Warning!!!

```
print(type(3.14))
print(3.14)

print(type(int(3.14)))
print(int(3.14))
```

- int to str : 정수를 문자로

```
print(type(9))
print(type(str(9)))
```

- float to str : 실수를 문자로

```
print(type(3.14))
print(type(str(3.14)))
```

> ## 2) Data Structure

- List to Tuple : 리스트를 튜플로

```
tuple1 = tuple([1,3,5,7,9])    #리스트 앞에 tuple이라고 쓰면 tuple 된다.
print(tuple1)
print(type(tuple1))     
```

- Tuple to List : 튜플을 리스트로

```
list0 = list((1,3,5,7,9))
print(list0)
print(type(list0))
```

- List to Dictionary : 리스트를 딕셔너리로

```
dict0 = dict([['A', 123], ['B', 234], ['C', 567]])
print(dict0)
print(type(dict0))
```
